```python
# Classify columns into categorical and numerical
def seprate_data_types(df):
    categorical = []
    numerical = []
    for column in df.columns:                # looping on the number of columns
        if df[column].nunique() < 100:       # nunique:  returns unique values in the column,
                                            #binary class(category) number of uniques will be 2
                                            #multi class(category) number of uniques will be < 100
                                            #here 100 is the hyperparamter and can be changed
            categorical.append(column)
        else:
            numerical.append(column)

    return categorical, numerical


categorical, numerical = seprate_data_types(df)         # Calling the function

# # Tabulate is a package used to print the list, dict or any data sets in a proper format; in table format
from tabulate import tabulate
table = [categorical, numerical]
print(tabulate({"Categorical":categorical,
                "continuous": numerical}, headers = ["categorical", "numerical"]))
```

```python
 #task1.1:Distribution of Special Requests
plt.figure(figsize=(12, 6))
sns.countplot(x='total_of_special_requests', data=df)
plt.title('Distribution of Number of Special Requests Made by Guests')
plt.xlabel('Number of Special Requests')
plt.ylabel('Count')
plt.show()
```

```python
#Interpretation:Most guests make between 0 and 2 special requests, with the majority making 0 requests. The number of guests making more than 3 special requests is significantly lower.
```

```python
#task1.2:
plt.figure(figsize=(12, 6))
sns.boxplot(x='total_of_special_requests', y='adr', data=df)
plt.title('ADR by Number of Special Requests')
plt.xlabel('Number of Special Requests')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()

# Calculate correlation
correlation = df[['total_of_special_requests', 'adr']].corr().iloc[0, 1]
print(f"Correlation between total special requests and ADR: {correlation:.2f}")

room_summary = df.groupby('reserved_room_type').agg(
    avg_adr=('adr', 'mean'),
    total_requests=('total_of_special_requests', 'sum'),
    count=('reserved_room_type', 'count')
).reset_index()

# Plotting room type preferences
plt.figure(figsize=(12, 6))
sns.barplot(x='reserved_room_type', y='avg_adr', data=room_summary)
plt.title('Average Daily Rate (ADR) by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()

# Total special requests by room type
plt.figure(figsize=(12, 6))
sns.barplot(x='reserved_room_type', y='total_requests', data=room_summary)
plt.title('Total Special Requests by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Total Special Requests')
plt.xticks(rotation=45)
plt.show()
```

```python
#Interpretation:Guests who make more special requests tend to have higher average daily rates (ADR). However, there is also a significant amount of variation within each group, especially for those with fewer requests. This suggests that other factors beyond the number of special requests influence ADR.
Room type A receives significantly more special requests than other room types. The remaining room types receive a relatively small number of special requests, with room types C, D, and H having the fewest. This suggests that room type A may be more popular or have features that attract more special requests.
```

```python
#task2.1
market_segment_summary = df.groupby('market_segment').agg(
    success_rate=('is_canceled', lambda x: (x == 0).mean()),  
    total_bookings=('is_canceled', 'count')
).reset_index()

channel_summary = df.groupby('distribution_channel').agg(
    success_rate=('is_canceled', lambda x: (x == 0).mean()),  
    total_bookings=('is_canceled', 'count')
).reset_index()

# Plotting bookings by market segment
plt.figure(figsize=(12, 6))
plt.subplot(121)  
sns.barplot(x='market_segment', y='total_bookings', data=market_segment_summary)
plt.title('Number of Bookings by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Total Bookings')
plt.xticks(rotation=45)

plt.subplot(122) 
sns.barplot(x='market_segment', y='success_rate', data=market_segment_summary)
plt.title('Success Rate by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()

# Plotting bookings by distribution channel 
plt.figure(figsize=(12, 6))
plt.subplot(121)
sns.barplot(x='distribution_channel', y='total_bookings', data=channel_summary)
plt.title('Number of Bookings by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Total Bookings')
plt.xticks(rotation=45)

plt.subplot(122)
sns.barplot(x='distribution_channel', y='success_rate', data=channel_summary)
plt.title('Success Rate by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#interpretation:Online TA (Online Travel Agencies) are the most popular market segment, receiving the most bookings. However, their success rate is relatively low compared to other segments. The corporate market segment, while having fewer bookings, boasts the highest success rate. This suggests that corporations are more likely to confirm their bookings once made, while Online TA bookings may be more prone to cancellations.
TA/TO (Travel Agents/Tour Operators) are the most popular distribution channel for bookings, but they also have the lowest success rate. Corporate bookings, while fewer in number, have the highest success rate. This suggests that corporate bookings are more reliable and less likely to be canceled compared to those through TA/TO. Direct bookings fall somewhere in between in terms of both booking volume and success rate.
```

```python
#task 2.2

df = df.dropna(subset=['meal', 'reserved_room_type', 'required_car_parking_spaces'])

# Group by meal and calculate average ADR
meal_adr = df.groupby('meal')['adr'].mean().sort_values()
plt.bar(meal_adr.index, meal_adr.values, color='skyblue')  # Use plt.bar for sorting
plt.title('Average Daily Rate by Meal Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()

# Group by room type and calculate average ADR
room_type_adr = df.groupby('reserved_room_type')['adr'].mean().sort_values()
plt.bar(room_type_adr.index, room_type_adr.values, color='lightgreen')  # Use plt.bar for sorting
plt.title('Average Daily Rate by Reserved Room Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()

# Group by parking spaces and calculate average ADR
parking_adr = df.groupby('required_car_parking_spaces')['adr'].mean()
plt.bar(parking_adr.index, parking_adr.values, color='salmon')  # Use plt.bar for sorting

# Adjust x-axis rotation based on number of categories (optional)
if len(parking_adr.index) > 5:
    plt.xticks(rotation=90)  # Rotate 90 degrees for many categories
else:
    plt.xticks(rotation=0)

plt.title('Average Daily Rate by Required Car Parking Spaces')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

```python
#interpretation:Average Daily Rate by Reserved Room Type: Room types with higher numbers (e.g., H) generally have higher ADRs compared to lower-numbered rooms. This suggests that room types with higher numbers might offer more features, amenities, or size, leading to higher prices.
Average Daily Rate by Required Car Parking Spaces: There's a slight increase in ADR for rooms requiring one car parking space compared to those requiring none. However, the difference is not substantial, indicating that the presence of car parking spaces might have a minimal impact on the overall room rate.
```

```python
#task2.3


# Distribution of Bookings Across Distribution Channels
plt.figure(figsize=(12, 6))
sns.countplot(x='distribution_channel', data=df)
plt.title('Distribution of Bookings Across Distribution Channels')
plt.xlabel('Distribution Channel')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()

# Average Daily Rate (ADR) by Distribution Channel
channel_adr = df.groupby('distribution_channel')['adr'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=channel_adr.index, y=channel_adr.values)
plt.title('Average Daily Rate (ADR) by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()

# Lead Time by Distribution Channel (Scatter Plot)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='lead_time', y='distribution_channel', hue='distribution_channel', data=df)
plt.title('Lead Time by Distribution Channel')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Distribution Channel')
plt.show()

```

```python
#Interpretation:the distribution of bookings across different distribution channels. TA/TO (Travel Agents/Tour Operators) are the most popular channel, receiving a significantly higher number of bookings compared to Direct, Corporate, and GDS (Global Distribution Systems). Direct bookings are the second most popular, while Corporate and GDS receive a very low number of bookings.
```

```python
#interpretation:The bar chart illustrates the average daily rate (ADR) for different distribution channels. Direct bookings have the highest ADR, followed by GDS and TA/TO. Corporate bookings have the lowest ADR. This suggests that customers booking directly or through GDS might be opting for higher-priced rooms or additional services, while corporate bookings might be for budget-conscious travelers or negotiated rates.
```

```python
#Interpretation:The scatter plot shows the distribution of lead times for different distribution channels. TA/TO (Travel Agents/Tour Operators) tend to have shorter lead times, while Corporate bookings often have longer lead times. Direct bookings fall somewhere in between. This suggests that corporate bookings might involve more complex processes or negotiations, leading to longer lead times. TA/TO, on the other hand, might have more streamlined booking processes or be able to process bookings more quickly.
```
