```python
#Display basic information about the dataset, which includes column names, number of rows and columns,and data type of the columns. Also, display the count of missing values for each column
#task1
import pandas as pd

def display_dataset_info(df):
    """
    Displays basic information about a pandas DataFrame.

    Args:
        df (pandas.DataFrame): The DataFrame to be analyzed.

    Returns:
        None
    """
    print("Column Names:")
    print(df.columns.tolist())
    print("\nNumber of Rows:", df.shape[0])
    print("Number of Columns:", df.shape[1])
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isna().sum())

# Example usage
data = {'Name': ['John', 'Jane', 'Bob', None, 'Alice'],
        'Age': [25, 30, None, 35, 28],
        'City': ['New York', 'London', 'Paris', 'Tokyo', None]}

df = pd.DataFrame(data)
display_dataset_info(df)
```

```python
#Separate the columns into categorical and numerical data types and display them
#task2.1
import pandas as pd

def separate_columns(df):
    """
    Separate the columns of a pandas DataFrame into categorical and numerical data types.
    
    Args:
        df (pandas.DataFrame): The input DataFrame.
        
    Returns:
        tuple: A tuple containing two lists:
            1. A list of column names for categorical data.
            2. A list of column names for numerical data.
    """
    categorical_cols = []
    numerical_cols = []
    
    for col in df.columns:
        if df[col].dtype == 'object':
            categorical_cols.append(col)
        else:
            numerical_cols.append(col)
    
    return categorical_cols, numerical_cols

# Example usage
data = {'Name': ['John', 'Jane', 'Bob', 'Alice'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'London', 'Paris', 'Tokyo']}

df = pd.DataFrame(data)

categorical_cols, numerical_cols = separate_columns(df)

print("Categorical Columns:")
for col in categorical_cols:
    print(col)

print("\nNumerical Columns:")
for col in numerical_cols:
    print(col)
```

```python
#Conduct descriptive analysis of numerical columns "PRICEEACH!, 'SALES', and "MSRP" to determine the unique values present in each column, their respective modes, and other relevant statistics.
#task2.2
import pandas as pd

def descriptive_analysis(data, columns):
    """
    Performs descriptive analysis on the specified columns of a pandas DataFrame.

    Args:
        data (pandas.DataFrame): The input DataFrame.
        columns (list): A list of column names to analyze.

    Returns:
        None (Prints the descriptive analysis results.)
    """
    for col in columns:
        print(f"Column: {col}")
        print(f"Unique Values: {data[col].unique()}")
        print(f"Mode: {data[col].mode()[0]}")
        print(f"Mean: {data[col].mean()}")
        print(f"Median: {data[col].median()}")
        print(f"Standard Deviation: {data[col].std()}")
        print(f"Minimum: {data[col].min()}")
        print(f"Maximum: {data[col].max()}")
        print("-" * 20)

# Example usage
data = pd.read_csv("C:/Users/91918/OneDrive/Desktop/SUDHEERA/DataScience with Gen Ai/Course5-EDA/Sprint4/DS1_C5_S3_Sales_Data_Practice.csv")
numerical_columns = ["PRICEEACH", "SALES", "MSRP"]
descriptive_analysis(data, numerical_columns)
```

```python
#Perform descriptive analysis of categorical columns "PRODUCTLINE','CUSTOMERNAME', and "COUNTRY" to identify the unique values present in each column, their respective modes, and other relevant statistics.
#task2.3
import pandas as pd

def descriptive_analysis(df):
    """
    Performs descriptive analysis on the specified columns in the given DataFrame.

    Args:
        df (pandas.DataFrame): The input DataFrame.

    Returns:
        None (Prints the descriptive statistics for each column)
    """
    columns = ['PRODUCTLINE', 'CUSTOMERNAME', 'COUNTRY']

    for col in columns:
        print(f"Column: {col}")
        print(f"Unique Values: {df[col].unique()}")
        print(f"Mode: {df[col].mode()[0]}")
        print(f"Frequency: \n{df[col].value_counts()}")
        print("-" * 20)

# Example usage
data = {
    'ORDERNUMBER': [1, 2, 3, 4, 5],
    'PRODUCTLINE': ['Sports', 'Sports', 'Fashion', 'Home', 'Sports'],
    'CUSTOMERNAME': ['John Doe', 'Jane Smith', 'Bob Johnson', 'Alice Williams', 'John Doe'],
    'COUNTRY': ['USA', 'UK', 'France', 'USA', 'UK']
}

df = pd.DataFrame(data)
descriptive_analysis(df)
```

```python
#Perform descriptive analysis of the numerical columns "PRICEEACH!, ‘SALES’, and MSRP" by using the following Bar chart, Pie chart, Measures of central tendency (mean, median), Standard deviation of the column, Skewness of the column
#task3
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('C:/Users/91918/OneDrive/Desktop/SUDHEERA/DataScience with Gen Ai/Course5-EDA/Sprint4/DS1_C5_S3_Sales_Data_Practice.csv')

# Descriptive analysis for 'PRICEEACH'
priceeach_stats = data['PRICEEACH'].describe()
print("Descriptive statistics for 'PRICEEACH':")
print(priceeach_stats)

# Plot a bar chart for 'PRICEEACH'
plt.figure(figsize=(8, 6))
plt.bar(data['PRODUCTLINE'], data['PRICEEACH'])
plt.xlabel('Product Line')
plt.ylabel('Price Each')
plt.title('Price Each by Product Line')
plt.xticks(rotation=45)
plt.show()

# Descriptive analysis for 'SALES'
sales_stats = data['SALES'].describe()
print("\nDescriptive statistics for 'SALES':")
print(sales_stats)

# Plot a pie chart for 'SALES'
plt.figure(figsize=(8, 6))
plt.pie(data['SALES'], labels=data['PRODUCTLINE'], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Sales Distribution by Product Line')
plt.show()

# Measures of central tendency (mean, median)
mean_priceeach = data['PRICEEACH'].mean()
median_priceeach = data['PRICEEACH'].median()
print("\nMean of 'PRICEEACH':", mean_priceeach)
print("Median of 'PRICEEACH':", median_priceeach)

mean_sales = data['SALES'].mean()
median_sales = data['SALES'].median()
print("\nMean of 'SALES':", mean_sales)
print("Median of 'SALES':", median_sales)

# Standard deviation of the column
priceeach_std = data['PRICEEACH'].std()
sales_std = data['SALES'].std()
print("\nStandard deviation of 'PRICEEACH':", priceeach_std)
print("Standard deviation of 'SALES':", sales_std)

# Skewness of the column
priceeach_skew = data['PRICEEACH'].skew()
sales_skew = data['SALES'].skew()
print("\nSkewness of 'PRICEEACH':", priceeach_skew)
print("Skewness of 'SALES':", sales_skew)
```

```python
#Perform descriptive analysis of the categorical columns 'PRODUCTLINE', 'CUSTOMERNAME', and ‘COUNTRY’ by using the following:Bar chart, Pie chart,Measures of central tendency (mean, median),Standard deviation of the column, Skewness of the column
#task4
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

def descriptive_analysis(df, columns):
    for col in columns:
        # Bar chart
        df[col].value_counts().plot(kind='bar')
        plt.title(f"Bar Chart for {col}")
        plt.show()

        # Pie chart
        df[col].value_counts().plot(kind='pie')
        plt.title(f"Pie Chart for {col}")
        plt.show()

        # Handle non-numeric columns gracefully
        if pd.api.types.is_numeric_dtype(df[col]):
            # Measures of central tendency
            mean = df[col].mean()
            median = df[col].median()
            print(f"Measures of central tendency for {col}:")
            print(f"Mean: {mean}")
            print(f"Median: {median}")

            # Standard deviation
            std = df[col].std()
            print(f"Standard deviation of {col}: {std}")

            # Skewness
            skew = stats.skew(df[col])
            print(f"Skewness of {col}: {skew}")
        else:
            print(f"Column '{col}' is non-numeric. Skipping calculations for this column.")

        print("-" * 20)

# Example usage
data = {
    'PRODUCTLINE': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'CUSTOMERNAME': ['John', 'Jane', 'Bob', 'Alice', 'Mike', 'Sarah', 'Tom', 'Emily', 'David', 'Emma'],
    'COUNTRY': ['USA', 'UK', 'France', 'Germany', 'China', 'India', 'Brazil', 'Canada', 'Australia', 'Japan']
}

df = pd.DataFrame(data)

descriptive_analysis(df, ['PRODUCTLINE', 'CUSTOMERNAME', 'COUNTRY'])
```
