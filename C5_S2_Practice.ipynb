```python
# Seprating out the categorical and continuous variable
def seprate_data_types(df):
    categorical = []
    numerical = []
    for column in df.columns:                # looping on the number of columns
        if df[column].nunique() < 100:       # nunique:  returns unique values in the column,
                                            #binary class(category) number of uniques will be 2
                                            #multi class(category) number of uniques will be < 100
                                            #here 100 is the hyperparamter and can be changed
            categorical.append(column)
        else:
            numerical.append(column)

    return categorical, numerical


categorical, numerical = seprate_data_types(df)         # Calling the function

# # Tabulate is a package used to print the list, dict or any data sets in a proper format; in table format
from tabulate import tabulate
table = [categorical, numerical]
print(tabulate({"Categorical":categorical,
                "continuous": numerical}, headers = ["categorical", "numerical"]))
```

```python
#Task3:
# Prepare the data: count number of attrited and non-attrited employees by job role
job_role_counts = df[df['Attrition'] == 'Yes'].groupby('JobRole').size().reset_index(name='Count')
job_role_counts = job_role_counts.sort_values(by='Count', ascending=True)

# Plotting
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x='Count', y='JobRole', data=job_role_counts, palette='viridis', ax=ax)
ax.set_xlabel('Count of Attrited Employees')
ax.set_ylabel('Job Role')
ax.set_title('Count of Attrited Employees by Job Role (Travel Rarely)')
plt.show()
```

```python
#interpretation:The most common job roles are Laboratory Technician, Sales Executive, and Research Scientist, while Research Director is the least common. The chart also shows the number of employees in each job role.
```

```python
#Task4:
# Filter for Life Sciences and Medical education backgrounds
filtered_df = df[df['EducationField'].isin(['Life Sciences', 'Medical'])]

# Calculate attrition counts by JobRole and EducationField
attrition_counts = filtered_df.groupby(['JobRole', 'Attrition']).size().unstack(fill_value=0)

# Add a column for total attrition count for each job role
attrition_counts['Total'] = attrition_counts.sum(axis=1)

# Sort by total attrition count
attrition_counts = attrition_counts.sort_values(by='Total', ascending=True)

# Plotting
fig, ax = plt.subplots(figsize=(10, 7))
attrition_counts[['Yes']].plot(kind='barh', ax=ax, color='skyblue', legend=False)
ax.set_xlabel('Count of Attrited Employees')
ax.set_ylabel('Job Role')
ax.set_title('Count of Attrited Employees by Job Role (Life Sciences and Medical Education)')
plt.show()
```

```python
#interpretation:
'''Laboratory Technician: This job role has the highest number of employees who have left the company.
Research Scientist: This job role also has a relatively high number of attriters.
Sales Executive: While the number of attriters is lower than in Laboratory Technician and Research Scientist, it's still significant.
Other Roles: The remaining job roles (Manufacturing Director, Healthcare Representative, Manager, Research Director, Sales Representative, and Human Resources) have a lower number of attriters'''
```
